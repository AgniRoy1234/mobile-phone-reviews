```python
def extract(n):
    pattern=r"(\d+) (\w+)"
    f=re.findall(pattern,n)
    return(f[0])
    
```

```python
extract("10 months")
```

```python
extract("5 days")
```

```python
dt=datetime.date(2022, 9, 21)
```

```python
def convert(n):
    tup=extract(n)
    num=int(float(tup[0]))
    t=tup[1]
    if(t=="months"):
        d=num*30
    elif(t=="month"):
        d=30
    elif(t=="days"):
        d=num
    elif(t=="day"):
        d=num
    return(dt-datetime.timedelta(days=d))
    
```

```python
convert("5 days")
```

```python
convert("10 months")
```
