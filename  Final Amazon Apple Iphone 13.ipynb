**AMAZON**

```python
dataset_amazon=[]
```

```python
for i in range(1,60):
  url=f"https://www.amazon.in/Apple-iPhone-13-128GB-Midnight/product-reviews/B09G9HD6PD/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}"
  browser.get(url)
  containers=[x.get_attribute('innerHTML') for x in browser.find_elements(By.XPATH,'//div[@class="a-section celwidget"]')]
  for container in containers:
    rev1=dict()
    bsObj=BeautifulSoup(container)
    #review date
    date=bsObj.find('span',attrs={"class":"a-size-base a-color-secondary review-date"}).text.strip()
    date=re.split("Reviewed in India on ",date)[-1]
    

    #review 
    rev=bsObj.find('span',attrs={"class":"a-size-base review-text review-text-content"}).text.strip()
    #rating
    try:
      rating=bsObj.find('a',attrs={"class":"a-link-normal"})["title"]
      rating=int(float(rating[0]))
    except:
      rating=np.NaN

    #found helpful
    try:
      helpful=bsObj.find('span',attrs={"class":"a-size-base a-color-tertiary cr-vote-text"}).text.strip()
      helpful=helpful.split(" ")[0]
      helpful=int(helpful)
    except:
      helpful=np.NaN

    #verfied purchase
    try:
      vpurchase=bsObj.find('span',attrs={"class":"a-size-base a-color-tertiary cr-vote-text"}).text.strip()
      vpurchase="Verified"
    except:
      vpurchase="Not Verified"

    #available info
    try:
      model=bsObj.find('a',attrs={"class":"a-size-mini a-link-normal a-color-secondary"}).text.strip()
      arr=re.split(r"(Colour: |Size name: |Style name: )",model)
      color=arr[4]
      size=arr[2]
    except:
      y="info not available"
      color=y
      size=y
      style=y

      
    rev1["date"]=date
    rev1["rating"]=rating
    rev1["review"]=rev
    rev1["found_helpful"]=helpful
    rev1["verified_purchase"]=vpurchase
    rev1["color"]=color
    rev1["size"]=size
    dataset_amazon.append(rev1)
```

```python
df_amazon=pd.DataFrame(dataset_amazon)
```
